```python
import pandas as pd
import uuid
import random
```

## Create Dummy DataFrame

```python
user_uuid = [str(uuid.uuid1()) for i in range(0,10_000)]

event_ids = [str(uuid.uuid1()) for i in range(0,1_000_000)]

category_ids = [round(random.random()*45) for i in range(0,1_000_000)]

user_ls = [user_uuid[round(random.random()*len(user_uuid))-1] for i in range(0,1_000_000)]

value_1 = [random.random() for i in range(0,1_000_000)]

value_2 = [random.random()*25 for i in range(0,1_000_000)]

value_3 = [random.random()*300 for i in range(0,1_000_000)]

df = pd.DataFrame({'user':user_ls,'event':event_ids,'category':category_ids,'a':value_1,'b':value_2,'c':value_3})

df
```

```python
from panda_searches import PandaSearches
```

## Indexing

```python
df.ps.indexing('user','event','category')
```

## Search Comparison

```python
test_user = df.user.sample(1).iloc[0] # Getting test user
test_user
```

```python
%%time

df[df['user']==test_user]
```

```python
%%time

df.ps[('user',test_user)]
```

## Search with AND/OR

```python
%%time

df[(df['user']==test_user) & (df['category']==2)]
```

```python
%%time

df.ps[ [ ('user',test_user),'and',('category',2) ] ]
```

## Complex Search

```python
test_user_list = list(df.user.sample(5))
test_user_list
```

```python
%%time

df[(df['user'].isin(test_user_list)) | ((df['user']==test_user) & (df['category']==2))]
```

```python
%%time

df.ps[ [ ('user',test_user_list),'or',[ ('user',test_user),'and',('category',2) ] ] ]
```
